```python
personas= [
    {"name": "Alice Johnson", "age": 28},
    {"name": "Bob Smith", "age": 34},
    {"name": "Charlie Davis", "age": 22},
    {"name": "Diana Prince", "age": 30}
]

# completar: sorted (personas, key=...)
mayor = max(personas, key=lambda p: p["edad"])
print("Ordenado por edad: ", mayor)

menor = min(personas, key=lambda p: p["edad"])
print("ordenado por nombre: ",menor)
```
