# Ejemplo de Carga de Datos
Cuaderno introductorio para verificar el entorno y practicar carga de datos con Pandas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Librerías cargadas correctamente.")
```

## 1. Crear un DataFrame de ejemplo

```python
# Crear datos de ejemplo
np.random.seed(42)
n = 100

datos = pd.DataFrame({
    'nombre': [f'Estudiante_{i}' for i in range(n)],
    'edad': np.random.randint(18, 30, n),
    'calificacion': np.random.uniform(60, 100, n).round(1),
    'asistencia': np.random.uniform(0.5, 1.0, n).round(2),
    'aprobado': np.random.choice([True, False], n, p=[0.7, 0.3])
})

datos.head(10)
```

## 2. Exploración básica

```python
print(f"Dimensiones: {datos.shape}")
print(f"\nTipos de datos:\n{datos.dtypes}")
print(f"\nEstadísticas descriptivas:")
datos.describe()
```

## 3. Visualización rápida

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Histograma de calificaciones
sns.histplot(datos['calificacion'], bins=15, kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('Distribución de Calificaciones')
axes[0].set_xlabel('Calificación')

# Scatter: calificación vs asistencia
sns.scatterplot(data=datos, x='asistencia', y='calificacion', hue='aprobado', ax=axes[1])
axes[1].set_title('Calificación vs Asistencia')

plt.tight_layout()
plt.show()
```

## 4. Guardar y cargar desde CSV

```python
# Guardar a CSV
datos.to_csv('datos_ejemplo.csv', index=False)
print('Archivo guardado: datos_ejemplo.csv')

# Cargar desde CSV
datos_cargados = pd.read_csv('datos_ejemplo.csv')
print(f'Datos cargados: {datos_cargados.shape[0]} filas, {datos_cargados.shape[1]} columnas')
datos_cargados.head()
```
